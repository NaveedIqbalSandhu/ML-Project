```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetics_prediction=pd.read_csv('C:\\Users\\IDEAL COM\\Downloads\\archive (29)\\diabetes.csv')
```

```python
x=diabetics_prediction.drop(columns='Outcome',axis=1)
y=diabetics_prediction['Outcome']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
standerddata=scaler.fit_transform(x)
```

```python
standerddata
```

```python
x=standerddata
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
classifier=svm.SVC(kernel='linear')
classifier.fit(x_train,y_train)
```

```python
x_train_prediction=classifier.predict(x_train)
accuracy_train_data=accuracy_score(x_train_prediction,y_train)
accuracy_train_data
```

```python
x_test_prediction=classifier.predict(x_test)
accuracy_test_data=accuracy_score(x_test_prediction,y_test)
accuracy_test_data
```

```python
input_data=[1,103,30,38,83,43.3,0.183,33]
input_data_numpy_array=np.asarray(input_data)

input_data_reshape=input_data_numpy_array.reshape(1,-1)
std_data=scaler.transform(input_data_reshape)

prediction=classifier.predict(std_data)

prediction
```
