```python
col=['Year', 'average_rain_fall_mm_per_year',
       'pesticides_tonnes', 'avg_temp','Area','Item','hg/ha_yield']
df=df[col]
df
```

```python
x=df.drop(columns='hg/ha_yield',axis=1)
y=df['hg/ha_yield']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
ohe=OneHotEncoder(drop='first')
scaler=StandardScaler()
```

```python
x_train.head(1)
```

```python
preprocessor=ColumnTransformer(
    transformers=[
        ('onehotencoder',ohe,[4,5]),
        ('standardScaler',scaler,[0,1,2,3])
    ],
    remainder='passthrough'
)
```

```python
preprocessor
```

```python
x_train_dummy=preprocessor.fit_transform(x_train)
x_test_dummy=preprocessor.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,r2_score
```

```python
models={
    'lr':LinearRegression(),
    'lss':Lasso(),
    'rg':Ridge(),
    'knn':KNeighborsRegressor(),
    'dtr':DecisionTreeRegressor()

}


for name, mod in models.items():
    mod.fit(x_train_dummy,y_train)
    y_pred=mod.predict(x_test_dummy)

    print(f'{name} MSE : {mean_squared_error(y_pred,y_test)} Score : {r2_score(y_pred,y_test)}')
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train_dummy,y_train)
dtr.predict(x_test_dummy)
```

```python
def prediction(Year,average_rain_fall_mm_per_year,pesticides_tonnes,avg_temp,Area,Item):
    features=np.array([[Year,average_rain_fall_mm_per_year,pesticides_tonnes,avg_temp,Area,Item]])

    transformed_features=preprocessor.transform(features)
    predicted_values=dtr.predict(transformed_features).reshape(1,-1)
    return predicted_values[0]
```

```python
Year=1990
average_rain_fall_mm_per_year=1485.0
pesticides_tonnes=121.0
avg_temp=16.37
Area='Albania'
Item='Maize'

result=prediction(Year,average_rain_fall_mm_per_year,pesticides_tonnes,avg_temp,Area,Item)
```
